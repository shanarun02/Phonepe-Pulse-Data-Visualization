# State Transaction Data

```python
def process_transaction_data(root_dir):
    data_list = []

    # Loop over all the state folders
    for state_dir in os.listdir(os.path.join(root_dir, r"D:\Phonepe\pulse\data\aggregated\transaction\country\india\state")):
        state_path = os.path.join(root_dir, r"D:\Phonepe\pulse\data\aggregated\transaction\country\india\state", state_dir)
        if os.path.isdir(state_path):
            # Loop over all the years folders
            for year_dir in os.listdir(state_path):
                year_path = os.path.join(state_path, year_dir)
                if os.path.isdir(year_path):
                    # Loop over all the JSON files (one for each quarter)
                    for json_file in os.listdir(year_path):
                        if json_file.endswith('.json'):
                            with open(os.path.join(year_path, json_file)) as f:
                                data = json.load(f)
                                # Extract the data we are interested in
                                for transaction_data in data['data']['transactionData']:
                                    payment_instruments = transaction_data.get('paymentInstruments', [])
                                    if payment_instruments:
                                        row_dict = {
                                            'states': state_dir,
                                            'Transaction_year': year_dir,
                                            'Quarters': int(json_file.split('.')[0]),
                                            'Transaction_Type': transaction_data['name'],
                                            'Transaction_Count': payment_instruments[0].get('count', 0),
                                            'Transaction_Amount': payment_instruments[0].get('amount', 0)
                                        }
                                        data_list.append(row_dict)

    state_trans_data = pd.DataFrame(data_list)
    return state_trans_data
root_directory = r"D:\Phonepe\pulse"
state_transaction_dataframe = process_transaction_data(root_directory)


```

```python
def create_transaction_dataframe(root_dir):
    data_list = []

    # Loop over all the year folders inside the india folder
    india_folder = os.path.join(root_dir, "data", "aggregated", "transaction", "country", "india")
    for year_dir in os.listdir(india_folder):
        year_path = os.path.join(india_folder, year_dir)
        if os.path.isdir(year_path):
            # Loop over all the JSON files (one for each quarter)
            for json_file in os.listdir(year_path):
                if json_file.endswith('.json'):
                    with open(os.path.join(year_path, json_file)) as f:
                        data = json.load(f)
                        # Extract the data we are interested in
                        for transaction_data in data['data']['transactionData']:
                            payment_instruments = transaction_data.get('paymentInstruments', [])
                            if payment_instruments:
                                row_dict = {
                                    'Transaction_year': year_dir,
                                    'Quarters': int(json_file.split('.')[0]),
                                    'Transaction_Type': transaction_data['name'],
                                    'Transaction_Count': payment_instruments[0].get('count', 0),
                                    'Transaction_Amount': payment_instruments[0].get('amount', 0)
                                }
                                data_list.append(row_dict)

    # Create a Pandas DataFrame from the list of dictionaries
    india_Trans_data = pd.DataFrame(data_list)

    # Return the DataFrame
    return india_Trans_data

# Usage example
root_directory = r"D:\Phonepe\pulse"
india_transaction_dataframe = create_transaction_dataframe(root_directory)
```

# India User Data

```python
import os
import json
import pandas as pd

def extract_user_data(root_dir):
    # Initialize an empty list to store the dictionaries
    data_list = []

    india_folder = os.path.join(root_dir, "india")

    for year_dir in os.listdir(india_folder):
        year_path = os.path.join(india_folder, year_dir)
        if os.path.isdir(year_path):
            for json_file in os.listdir(year_path):
                if json_file.endswith('.json'):
                    json_path = os.path.join(year_path, json_file)
                    with open(json_path) as f:
                        data = json.load(f)
                        quarter = int(json_file.split('.')[0])

                        # Check if 'usersByDevice' is not None
                        if data['data'].get('usersByDevice') is not None:
                            # Extract relevant details and append to the list
                            for device_data in data['data']['usersByDevice']:
                                brand = device_data.get('brand', 'N/A')
                                count = device_data.get('count', 0)

                                row_dict = {
                                    'Year': year_dir,
                                    'Quarter': quarter,
                                    'Brand': brand,
                                    'Count': count
                                }
                                data_list.append(row_dict)

    # Create a Pandas DataFrame from the list of dictionaries
    user_overall_data = pd.DataFrame(data_list)

    return user_overall_data

# Usage example
root_directory = r"D:\Phonepe\pulse\data\aggregated\user\country"
india_user_data = extract_user_data(root_directory)
```

# Map India Transaction Data

```python
def extract_map_transaction_data(root_dir):
    # Initialize an empty list to store the dictionaries
    data_list = []

    for year_dir in os.listdir(root_dir):
        year_path = os.path.join(root_dir, year_dir)
        if os.path.isdir(year_path):
            for quarter_file in os.listdir(year_path):
                if quarter_file.endswith('.json'):
                    quarter_path = os.path.join(year_path, quarter_file)
                    with open(quarter_path) as f:
                        data = json.load(f)
                        quarter = int(quarter_file.split('.')[0])

                        for entry in data.get('data', {}).get('hoverDataList', []):
                            name = entry.get('name', 'N/A')
                            metric = entry.get('metric', [{}])[0]
                            count = metric.get('count', 0)
                            amount = metric.get('amount', 0)

                            # Append the data to the list
                            row_dict = {
                                'Year': year_dir,
                                'Quarter': quarter,
                                'Name': name,
                                'Count': count,
                                'Amount': amount
                            }
                            data_list.append(row_dict)

    # Create a Pandas DataFrame from the list of dictionaries
    map_trans_india = pd.DataFrame(data_list)

    return map_trans_india


root_directory = r"D:\Phonepe\pulse\data\map\transaction\hover\country\india"
india_map_trans = extract_map_transaction_data(root_directory)
```

# map state transaction data

```python
def extract_top_india_state_data(root_dir):
    # Create an empty list to store DataFrames
    dfs = []

    for year_dir in os.listdir(root_dir):
        year_path = os.path.join(root_dir, year_dir)
        if os.path.isdir(year_path):
            for quarter_file in os.listdir(year_path):
                if quarter_file.endswith('.json'):
                    quarter_path = os.path.join(year_path, quarter_file)
                    with open(quarter_path) as f:
                        data = json.load(f)
                        quarter = int(quarter_file.split('.')[0])

                        # Extract and append data to the list
                        state_data_list = []
                        for state_data in data.get('data', {}).get('states', []):
                            entity_name = state_data.get('entityName', 'N/A')
                            metric = state_data.get('metric', {})
                            count = metric.get('count', 0)
                            amount = metric.get('amount', 0)

                            state_data_list.append({
                                'Year': int(year_dir),
                                'Quarter': quarter,
                                'EntityName': entity_name,
                                'Count': count,
                                'Amount': amount
                            })

                        # Create a DataFrame from the list
                        state_data_df = pd.DataFrame(state_data_list)

                        # Append the DataFrame to the list
                        dfs.append(state_data_df)

    # Concatenate all DataFrames in the list into a single DataFrame
    top_india_state_data = pd.concat(dfs, ignore_index=True)

    return top_india_state_data

# Usage example
root_directory = r"D:\Phonepe\pulse\data\top\transaction\country\india"
top_india_state_data = extract_top_india_state_data(root_directory)
```
